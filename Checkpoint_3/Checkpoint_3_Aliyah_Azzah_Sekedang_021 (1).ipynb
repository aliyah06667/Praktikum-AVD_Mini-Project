### 2. Pengecekan Data

#### **Missing Values**

```python
print((df.isna().sum() / len(df)) * 100)
```

Dalam dataset ini, terdapat missing values pada kolom Satisfaction Level sebesar 0,57%.

Cara Menangani Missing Values:
Jika >= 70%, data biasanya dihapus karena terlalu banyak yang hilang.
Jika < 70%, ada dua pilihan:
Dihapus (jika jumlahnya kecil dan tidak berpengaruh besar).
Diimputasi (diisi ulang dengan nilai tertentu, seperti mean, median, modus, atau quartil tergantung jenis dan distribusi data).
Karena persentasenya kecil (<70%) dan data ini penting, saya memilih tidak menghapusnya, melainkan melakukan imputasi.

Metode yang digunakan adalah imputasi berdasarkan modus, karena Satisfaction Level merupakan data kategorikal, sehingga tidak bisa diisi menggunakan mean atau median.

```python
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("\nMissing Values Setelah Imputasi:\n", df.isnull().sum())
```

Tabel ini menunjukkan jumlah missing values (nilai yang hilang) dalam setiap kolom setelah dilakukan imputasi (pengisian nilai yang hilang). Semua kolom memiliki nilai 0, yang berarti tidak ada lagi data yang hilang dalam dataset setelah proses imputasi.
Ini menandakan bahwa teknik imputasi yang digunakan berhasil menggantikan semua nilai yang kosong, sehingga dataset sudah bersih dan siap untuk dianalisis lebih lanjut.

#### **Duplikat**

```python
df[df.duplicated()]
```

Setelah hasil pengecekan data duplikat. Saat dicek, tidak ada baris atau data yang muncul, artinya tidak ada nilai yang terduplikat dalam dataset. Karena itu, tidak perlu dilakukan langkah tambahan untuk menangani duplikasi.

#### **Outliers**

```python
results = []

numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_cols:
    q1 = df[col].dropna().quantile(0.25)
    q3 = df[col].dropna().quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    percent_outliers = (len(outliers) / len(df)) * 100

    results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)

display(results_df)
```

Setelah Pengecekan Outliers diatas, dapat terlihat bahwa persentase outliers dari tiap kolom yang bertipe numerik memiliki nilai 0.0 atau tidak terdapat outliers dari tiap data, sehingga tidak perlu dilakukan penanganan terhadap Outliers.

#### **Construct Data**

```python
df["Days Since Last Purchase"] = pd.to_numeric(df["Days Since Last Purchase"], errors="coerce")

df["Last Purchase Date"] = pd.Timestamp.today() - pd.to_timedelta(df["Days Since Last Purchase"], unit='D')

df["Day"] = df["Last Purchase Date"].dt.day_name()

df["Total Spend"] = pd.to_numeric(df["Total Spend"], errors="coerce")
df["Items Purchased"] = pd.to_numeric(df["Items Purchased"], errors="coerce")

df["Spend_per_Item"] = df["Total Spend"] / df["Items Purchased"]
df["Spend_per_Item"].fillna(0, inplace=True)

median_spend = df["Total Spend"].median()
df["High Value Customer"] = df["Total Spend"] > median_spend

df["Satisfied"] = df["Satisfaction Level"] == "Satisfied"

print(df.head())
```

#### **Data Reduction**

```python
df.drop(columns=["Customer ID", "Last Purchase Date"], inplace=True)

df = df[df["Days Since Last Purchase"] <= 30]

df["Membership Category"] = df["Membership Type"].replace({
    "Gold": "Premium",
    "Silver": "Premium",
    "Bronze": "Regular"
})

print(df.head())
```

**Komposisi**

```python
satisfaction_comp = pd.crosstab(df['Membership Type'], df['Satisfaction Level'])
satisfaction_comp_percent = satisfaction_comp.div(satisfaction_comp.sum(axis=1), axis=0) * 100

plt.figure(figsize=(6, 3))
ax = satisfaction_comp_percent.plot(
    kind='bar', stacked=True,
    colormap='OrRd'
)

for container in ax.containers:
    ax.bar_label(container, labels=[f'{x:.1f}%' if x >= 5 else '' for x in container.datavalues],
                 label_type='center', fontsize=9, color='white', weight='bold')

plt.title('Komposisi Tingkat Kepuasan berdasarkan Jenis Keanggotaan')
plt.xlabel('Jenis Keanggotaan')
plt.ylabel('Persentase (%)')
plt.xticks(rotation=0)
plt.legend(title='Tingkat Kepuasan')
plt.show()
```
