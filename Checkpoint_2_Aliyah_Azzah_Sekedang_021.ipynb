#### **Outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

```python
df
```

Melakukan imputasi

```python
columns_to_impute = ["Satisfaction Level"]

for col in columns_to_impute:
    mode_value = df[col].mode()[0]

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].fillna(mode_value)
```

#### **Construct Data**

```python
df["Days Since Last Purchase"] = pd.to_numeric(df["Days Since Last Purchase"], errors="coerce")

df["Last Purchase Date"] = pd.Timestamp.today() - pd.to_timedelta(df["Days Since Last Purchase"], unit='D')

df["Day"] = df["Last Purchase Date"].dt.day_name()

df["Total Spend"] = pd.to_numeric(df["Total Spend"], errors="coerce")
df["Items Purchased"] = pd.to_numeric(df["Items Purchased"], errors="coerce")

df["Spend_per_Item"] = df["Total Spend"] / df["Items Purchased"]
df["Spend_per_Item"].fillna(0, inplace=True)

median_spend = df["Total Spend"].median()
df["High Value Customer"] = df["Total Spend"] > median_spend

df["Satisfied"] = df["Satisfaction Level"] == "Satisfied"

print(df.head())
```

#### **Data Reduction**

```python
df.drop(columns=["Customer ID", "Last Purchase Date"], inplace=True)

df = df[df["Days Since Last Purchase"] <= 30]

df["Membership Category"] = df["Membership Type"].replace({
    "Gold": "Premium",
    "Silver": "Premium",
    "Bronze": "Regular"
})

print(df.head())
```
